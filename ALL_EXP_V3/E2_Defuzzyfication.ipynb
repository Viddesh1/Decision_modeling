```python
# Note:- 
# 1) For Trapizoidal membership function kindly use "trapmf" instead of trimf which is triangular membership function.
# 2) trimf() takes 4 parameters and trimf() takes 3

# 1) Centroid method

import matplotlib.pyplot as plt
import numpy as np
import skfuzzy as fuzz

# Define the triangular membership function
x = np.arange(0, 41)
mfx = fuzz.trimf(x, [10, 20, 30])

# For Trapizoidal membership function
# mfx = fuzz.trapmf(x, [10, 20, 30, 40])

# Plot the membership function
plt.plot(x, mfx, 'b', linewidth=1.5, label='Fuzzy set')

# Calculate the centroid using skfuzzy
centroid = fuzz.defuzz(x, mfx, 'centroid')

# Plot the centroid
plt.axvline(x=centroid, color='k', linestyle='--')
# plt.fill_betweenx([0, np.interp(centroid, x, mfx)], [0, centroid], alpha=0.1)

# Add labels and legends
plt.title("Centroid Method")
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Membership degree')
plt.legend()

# Show the plot
plt.show()

print('Centroid: {:.2f}'.format(centroid))
```

```python
# 2) Weighted Average Method

import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Define the input variable
x = np.arange(0, 41)

# Define the membership function
mfx = fuzz.trimf(x, [10, 20, 30])

# For Trapizoidal membership function
# mfx = fuzz.trapmf(x, [10, 20, 30, 40])

# Apply the weighted average defuzzification method
centroid_wa = np.sum(x * mfx) / np.sum(mfx)

# Plot the membership function
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(x, mfx, 'b', linewidth=1.5, label='Membership Function')
ax.fill_between(x, 0, mfx, alpha=0.1)

# Plot the centroid
ax.axvline(x=centroid_wa, color='r', linestyle='--', linewidth=1.5, label='Centroid (WA)')
ax.legend()

# Set plot attributes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
ax.set_title('Weighted Average Defuzzification')
ax.set_xlabel('Input Variable (x)')
ax.set_ylabel('Membership Degree')
plt.show()

```

```python
# 3) Center of Sum method

import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Define the input variable
x = np.arange(0, 41)

# Define the membership function for traingular function
# mfx = fuzz.trimf(x, [10, 20, 30])

# For Trapizoidal membership function
mfx = fuzz.trapmf(x, [10, 20, 30, 40])

# Apply the center of sum defuzzification method
centroid_cos = np.sum(mfx * (x + np.random.normal(size=x.size))) / np.sum(mfx)

# Plot the membership function
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(x, mfx, 'b', linewidth=1.5, label='Membership Function')
ax.fill_between(x, 0, mfx, alpha=0.1)

# Plot the centroid
ax.axvline(x=centroid_cos, color='r', linestyle='--', linewidth=1.5, label='Centroid (COS)')
ax.legend()

# Set plot attributes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
ax.set_title('Center of Sum Defuzzification')
ax.set_xlabel('Input Variable (x)')
ax.set_ylabel('Membership Degree')
plt.show()
```

```python
# 4) Mean of Maximum method

import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Define the input variable
x = np.arange(0, 41)

# Define the membership function
# mfx = fuzz.trimf(x, [10, 20, 30])

# For Trapizoidal membership function
mfx = fuzz.trapmf(x, [10, 20, 30, 40])

# Apply the mean of maximum defuzzification method
centroid_mom = fuzz.defuzz(x, mfx, 'mom')

# Plot the membership function
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(x, mfx, 'b', linewidth=1.5, label='Membership Function')
ax.fill_between(x, 0, mfx, alpha=0.1)

# Plot the centroid
ax.axvline(x=centroid_mom, color='r', linestyle='--', linewidth=1.5, label='Centroid (MOM)')
ax.legend()

# Set plot attributes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
ax.set_title('Mean of Maximum Defuzzification')
ax.set_xlabel('Input Variable (x)')
ax.set_ylabel('Membership Degree')
plt.show()
```
