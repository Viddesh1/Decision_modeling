### Monte-Carlo Simulation

Estimating the value of pi (π) using Monte-Carlo Simulation Method.

Let's say we want to estimate the value of π (pi) using Monte Carlo simulation. We can do this by imagining a circle inside a square, and then randomly throwing darts at the square. The idea is that the proportion of darts that land inside the circle should be proportional to the area of the circle divided by the area of the square, which is π/4.

Here's a simple visualization of this process:

1) Draw a square and a circle inside it, both with the same center:

2) Generate a large number of random points within the square by randomly choosing an x and y coordinate between -1 
and 1.

3) For each point, determine if it falls inside the circle or outside the circle by calculating the distance between the point and the center of the circle. If the distance is less than or equal to the radius of the circle (which is 1 in this case), then the point falls inside the circle.

4) Calculate the proportion of points that fall inside the circle to the total number of points generated. This proportion should be approximately π/4.

```python
# Without Visualization - Simple Monte-carlo simulation for approximation of PI
import random

n = 1000000  # number of points to generate
count = 0

# generate n random points in a 1x1 square
for i in range(n):
    x = random.uniform(0, 1)
    y = random.uniform(0, 1)
    
    # check if the point lies inside the unit circle
    if x**2 + y**2 <= 1:
        count += 1

# approximate pi using the ratio of points inside the unit circle
pi_approx = 4 * count / n

print("Approximation of pi:", pi_approx)
```

```python
# Simple Example of Monte-Carlo Simulation.
import random
import matplotlib.pyplot as plt

# Set the number of points to generate
num_points = 10000

# Generate random points within the square
points = []
for i in range(num_points):
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)
    points.append((x, y))

# Calculate the distance from each point to the center of the circle
inside_circle = []
outside_circle = []
for point in points:
    x, y = point
    distance = (x**2 + y**2)**0.5
    if distance <= 1:
        inside_circle.append(point)
    else:
        outside_circle.append(point)

# Calculate the value of pi
pi_estimate = 4 * len(inside_circle) / len(points)

# Plot the results
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.scatter([p[0] for p in inside_circle], [p[1] for p in inside_circle], color='blue', s=1)
ax.scatter([p[0] for p in outside_circle], [p[1] for p in outside_circle], color='red', s=1)
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.set_title(f'Estimate of π: {pi_estimate:.3f}')
plt.show()
```
